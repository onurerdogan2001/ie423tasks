### Statistical Tests

I want to do a couple of tests for my suspects. I have guessed that these factors (Month, ProductRelatedDuration, VisitorType and PageValue) effect the Purchase really well.

```python
from scipy.stats import chi2_contingency

# Contingency table for Month and Revenue
contingency_table_month = pd.crosstab(df['Month'], df['Revenue'])

# Chi-squared test
chi2, p, dof, ex = chi2_contingency(contingency_table_month)
print(f"Chi-squared test for Month vs. Revenue: p-value = {p}")
```

The p-value is low, thus it can be said revenue and month is related.

```python
from scipy.stats import ttest_ind

# Split ProductRelated_Duration into two groups based on Revenue
purchase_duration = df[df['Revenue'] == True]['ProductRelated_Duration']
no_purchase_duration = df[df['Revenue'] == False]['ProductRelated_Duration']

# t-test
t_stat, p_val = ttest_ind(purchase_duration, no_purchase_duration)
print(f"t-test for ProductRelated_Duration vs. Revenue: p-value = {p_val}")
```

The p-value is low, thus it can be said revenue and ProductRelatedDuration is related.

```python
# Contingency table for VisitorType and Revenue
contingency_table_visitor = pd.crosstab(df['VisitorType'], df['Revenue'])

# Chi-squared test
chi2, p, dof, ex = chi2_contingency(contingency_table_visitor)
print(f"Chi-squared test for VisitorType vs. Revenue: p-value = {p}")
```

The p-value is low, thus it can be said revenue and visitortype is related.

```python
# Split ProductRelated_Duration into two groups based on Revenue
p1 = df[df['Revenue'] == True]['PageValues']
p2 = df[df['Revenue'] == False]['PageValues']

# t-test
t_stat, p_val = ttest_ind(p1, p2)
print(f"t-test for PageValues vs. Revenue: p-value = {p_val}")
```

## Feature Engineering

### Creating New Features

I want to explore the data by creating new features.

I already created these:ProductRelated_Duration_Binned and PageValues_Binary.

1.   First is investigating the average duration time on the product related pages.

2.   BounceRate * ExitRate can show that if the value is high the page is not doing well.

3. Total Pages visited can be helpfull to see if it affects the Purchase

4. Average Duration on total pages can be helpfull too.




```python
# Interaction Features
df['ProductRelated_AvgDuration'] = df['ProductRelated_Duration'] / df['ProductRelated']
df['ProductRelated_AvgDuration'].replace([float('inf'), -float('inf')], 0, inplace=True)  # Replace infinities with 0
df['ProductRelated_AvgDuration'].fillna(0, inplace=True)  # Replace NaN values with 0

df['Bounce_Exit_Interaction'] = df['BounceRates'] * df['ExitRates']

# Aggregated Features
df['TotalPages'] = df['Administrative'] + df['Informational'] + df['ProductRelated']
df['AvgDuration'] = (df['Administrative_Duration'] + df['Informational_Duration'] + df['ProductRelated_Duration']) / df['TotalPages']
df['AvgDuration'].replace([float('inf'), -float('inf')], 0, inplace=True)  # Replace infinities with 0
df['AvgDuration'].fillna(0, inplace=True)  # Replace NaN values with 0
```

Since many of them is related with ProductRelated (and ProductRelated_Duration, they have correlation too), I need to be careful about redundancy.

```python
# Display the new features created
new_features = [
    'ProductRelated_AvgDuration',
    'Bounce_Exit_Interaction',
    'TotalPages',
    'AvgDuration',
    'PageValues_Binary'
]

# Display the first few rows of the new features
df[new_features].head()
```

So, Many of the new features are highly right-skewed. To improve this situation, I will try log transformations.

```python
selected_features = [
    'Month',
    'VisitorType',
    'Weekend',
    'TotalPages_log',
    'PageValues_Binary',
    'Bounce_Exit_Interaction',
    'ProductRelated_AvgDuration_log',
    'TrafficType',
    'Region',
    'Browser',
    'OperatingSystems',
    'SpecialDay'
]

X = df[selected_features]
y = df['Revenue']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the first few rows of the prepared data
print(X_train.head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Train Logistic Regression
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)

# Train Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Train Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)

from sklearn.tree import DecisionTreeClassifier

# Train Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

from sklearn.svm import SVC

# Train SVM
svm_model = SVC(random_state=42, probability=True)
svm_model.fit(X_train, y_train)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Predictions
y_pred_lr = lr_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)
y_pred_gb = gb_model.predict(X_test)
y_pred_dt = dt_model.predict(X_test)
y_pred_svm = svm_model.predict(X_test)

# Evaluation function
def evaluate_model(y_test, y_pred):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    return accuracy, precision, recall, f1, roc_auc

# Logistic Regression Evaluation
lr_results = evaluate_model(y_test, y_pred_lr)
print(f"Logistic Regression: Accuracy = {lr_results[0]}, Precision = {lr_results[1]}, Recall = {lr_results[2]}, F1-score = {lr_results[3]}, ROC-AUC = {lr_results[4]}")

# Random Forest Evaluation
rf_results = evaluate_model(y_test, y_pred_rf)
print(f"Random Forest: Accuracy = {rf_results[0]}, Precision = {rf_results[1]}, Recall = {rf_results[2]}, F1-score = {rf_results[3]}, ROC-AUC = {rf_results[4]}")

# Gradient Boosting Evaluation
gb_results = evaluate_model(y_test, y_pred_gb)
print(f"Gradient Boosting: Accuracy = {gb_results[0]}, Precision = {gb_results[1]}, Recall = {gb_results[2]}, F1-score = {gb_results[3]}, ROC-AUC = {gb_results[4]}")

# Decision Tree Evaluation
dt_results = evaluate_model(y_test, y_pred_dt)
print(f"Decision Tree: Accuracy = {dt_results[0]}, Precision = {dt_results[1]}, Recall = {dt_results[2]}, F1-score = {dt_results[3]}, ROC-AUC = {dt_results[4]}")

# SVM Evaluation
svm_results = evaluate_model(y_test, y_pred_svm)
print(f"SVM: Accuracy = {svm_results[0]}, Precision = {svm_results[1]}, Recall = {svm_results[2]}, F1-score = {svm_results[3]}, ROC-AUC = {svm_results[4]}")
```
