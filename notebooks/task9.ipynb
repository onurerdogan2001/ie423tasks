### BERT

```python
from transformers import pipeline

# Load a pre-trained BERT model for NER
ner_pipeline = pipeline('ner', model='dbmdz/bert-large-cased-finetuned-conll03-english', grouped_entities=True)

# Example text
text = "Apple is looking at buying U.K. startup for $1 billion."

# Perform NER
ner_results = ner_pipeline(text)

# Print results
for entity in ner_results:
    print(f"Entity: {entity['word']}, Label: {entity['entity_group']}, Score: {entity['score']}")
```

```python
text = "Galatasaray just lost the final match to the Beşiktaş in a humiliated way."

# Perform NER
ner_results = ner_pipeline(text)

# Print results
for entity in ner_results:
    print(f"Entity: {entity['word']}, Label: {entity['entity_group']}, Score: {entity['score']}")
```

### BART

```python
# Load a pre-trained BART model for summarization
summarizer = pipeline('summarization', model='facebook/bart-large-cnn')

# Example text
text = """
Machine learning is the study of computer algorithms that can improve automatically through experience and by the use of data.
It is seen as a part of artificial intelligence. Machine learning algorithms build a model based on sample data, known as training data,
in order to make predictions or decisions without being explicitly programmed to do so.
"""

# Perform summarization
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

# Print summary
print(summary[0]['summary_text'])
```

### Zero Shot Classification

```python
# Load a pre-trained model for zero-shot classification
classifier = pipeline('zero-shot-classification')

# Example text and candidate labels
text = "I am tired of tasks!"
candidate_labels = ["School", "education", "homework"]

# Perform classification
result = classifier(text, candidate_labels)

# Print results
print(result)
```

```python
# Example text and candidate labels
text = "I am going to graduate soon"
candidate_labels = ["School", "Education", "University"]

# Perform classification
result = classifier(text, candidate_labels)

# Print results
print(result)
```

### Translator

```python
from transformers import pipeline

# Load a pre-trained model for translation
translator = pipeline('translation_en_to_fr')

# Example text
text = "I am tired of tasks."

# Perform translation
translation = translator(text, max_length=40)

# Print translation
print(translation[0]['translation_text'])
```
